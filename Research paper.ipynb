## Abstract

The COVID-19 pandemic continues to have a devastating effect on the health and well-being of the global population. A critical step in the fight against COVID-19 is effective screening of infected patients, with one of the key screening approaches being radiology examination using chest radiography. It was found in early studies that patients present abnormalities in chest radiography images that are characteristic of those infected with COVID-19. 
Motivated by this, in this study is introduced COVID-Net, a deep convolutional neural network design tailored for the detection of COVID-19 cases from chest X-ray (CXR) images that is open source and available to the general public.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.graph_objects as go
import plotly.offline
import pandas as pd
```

```python
test_count = ["Results between Test and Train", "Test", "Train"]
test_count_values = [("Normal", 7966, 885),
                    ("Pneumonia", 5457, 594),
                    ("COVID-19", 2158, 291)]

df = pd.DataFrame(data=test_count_values, columns=test_count)
df.plot.bar(x='Results between Test and Train')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```
